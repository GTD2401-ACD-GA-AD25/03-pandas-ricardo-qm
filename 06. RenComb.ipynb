```python
canadian_youtube = pd.read_csv("https://raw.githubusercontent.com/GTD2401-ACD-GA-AD25/ACD.datasets.kaggle/refs/heads/main/CAvideos.txt")
british_youtube = pd.read_csv("https://raw.githubusercontent.com/GTD2401-ACD-GA-AD25/ACD.datasets.kaggle/refs/heads/main/GBvideos.txt")

pd.concat([canadian_youtube, british_youtube])
```

```python
left = canadian_youtube.set_index(['title', 'trending_date'])
```

```python
right = british_youtube.set_index(['title', 'trending_date'])
```
